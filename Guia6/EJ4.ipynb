```python
import pandas as pd
import matplotlib.pyplot as plt
#importo el archivo, que contiene la poblacion (en cientos de miles) de las 10 ciudades mas pobladas de cada pais
#en el año 1967.
df = pd.read_csv("databases/ciudades.txt",delimiter='\s+')
df
```

a) Construir en paralelo, para facilitar la comparación, un boxplot para los datos de
cada país e identificar los puntos extremos en cada uno de ellos

```python
# Construyo el boxplot para los datos de cada país
arg = df.boxplot(column=['Argentina'])
arg.set_title("Argentina")
arg.set_ylabel("Población (en cientos de miles)")
plt.show(arg)

```

```python
# Construyo el boxplot para los datos de cada país
usa = df.boxplot(column=['EEUU'])
usa.set_title("EEUU")
usa.set_ylabel("Población (en cientos de miles)")
plt.show(usa)
```
