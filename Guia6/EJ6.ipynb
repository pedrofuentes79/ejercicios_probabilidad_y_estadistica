```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import probplot
import scipy.stats as stats
```

a)
Generar muestras de tamaño 25, 50 y 100 de una variable con distribución normal. Construir QQ-plots para cada una de ellas. Repetir con diferentes semillas.

```python
# Seteo la semilla, al cambiarla obtenemos distintos resultados en los qq plots del ej a)
np.random.seed(14)
```

```python
# Genero muestras de tamaño 25 de una variable con distribución normal de media 0 y desvio 1
muestra_25 = np.random.normal(0, 1, 25)
muestra_50 = np.random.normal(0, 1, 50)
muestra_100 = np.random.normal(0, 1, 100)

# Generlo los qq plots para  cada una
probplot(muestra_25, plot=plt)
plt.title("QQ Plot para muestra de tamaño 25")
plt.show()
```

```python
# Genero qq plot para la muestra de 50
probplot(muestra_50, plot=plt)
plt.title("QQ Plot para muestra de tamaño 50")
plt.show()
```

```python
# Genero el qq plot para la muestra de 100
probplot(muestra_100, plot=plt)
plt.title("QQ Plot para muestra de tamaño 100")
plt.show()
```

b)
Repetir a) para una variable con distribución Γ(5, 1/2).

```python
# Genero muestras de tamaño 25, 50 y 100 de una variable gamma 5, 0.5
muestra_gamma_25 = np.random.gamma(5, 0.5, 25)
muestra_gamma_50 = np.random.gamma(5, 0.5, 50)
muestra_gamma_100 = np.random.gamma(5, 0.5, 100)    
```

```python
# Genero los qq plots para la muestra de 25
probplot(muestra_gamma_25, dist=stats.gamma, sparams=(5, 0.5), plot=plt)
plt.title("QQ Plot para muestra de tamaño 25")
plt.show()
```

```python
# Genero los qq plots para la muestra de 50
probplot(muestra_gamma_50, dist=stats.gamma, sparams=(5, 0.5), plot=plt)
plt.title("QQ Plot para muestra de tamaño 50")
plt.show()
```

```python
# Genero el qq plot para la muestra de 100
probplot(muestra_gamma_100, dist=stats.gamma, sparams=(5, 0.5), plot=plt)
plt.title("QQ Plot para muestra de tamaño 100")
plt.show()

```

c)
Repetir a) para Y = Z * U donde Z ∼ N(0,1) y U ∼ U(0,1) independientes.

```python
z_25 = np.random.normal(0, 1, 25)
u_25 = np.random.uniform(0, 1, 25)
y_25 = z_25 * u_25

z_50 = np.random.normal(0, 1, 50)
u_50 = np.random.uniform(0, 1, 50)
y_50 = z_50 * u_50

z_100 = np.random.normal(0, 1, 100)
u_100 = np.random.uniform(0, 1, 100)
y_100 = z_100 * u_100
```

```python
# Genero los qq plots para la muestra de 25
probplot(y_25, plot=plt)
plt.title("QQ Plot para muestra de tamaño 25")
plt.show()
```
