```python
import pandas as pd
```

```python
df = pd.read_csv("databases/lamparas.txt", sep="\t", header=None)
df.columns = ["Vida"]
```

```python
from scipy.stats import trim_mean

print("Media muestral: ", df["Vida"].mean())
print("Media muestral 0.1-podada", trim_mean(df["Vida"], 0.1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Asumo que la función de distribución empírica es la función de distribución acumulada
# Calcular la función de distribución empírica  
x = np.sort(df)
y = np.arange(1, len(df) + 1) / len(df)

# Graficar la función de distribución empírica
# plt.step() se usa para trazar la función de distribución empírica como una gráfica escalonada.
plt.step(x, y, where='post')
plt.xlabel('Vida útil de las lámparas')
plt.ylabel('Probabilidad acumulada')
plt.title('Función de distribución empírica')
plt.show()
```
